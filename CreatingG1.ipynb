```python
#importing the libraries for the graph
import networkx as nx

#creating an empty graph
DG = nx.DiGraph()

#Adding nodes
edges = [("Mother","Father"),("Mother", "Daughter"),("Daughter's Friend", "Father"), ("Daughter's Friend", "Homeroom_Teacher"), ("Mother", "Daughter's Friend Teacher"), ("Father", "Daughter's Friend Teacher")]
DG.add_edges_from(edges)

#Customize the network

nx.draw(
    DG,
    node_color="green",
    node_size=100
)
```
